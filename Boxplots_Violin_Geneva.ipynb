```r
install.packages("viridis")
library("viridis")
install.packages("missMDA")
library("missMDA")
install.packages("factoextra")
library("factoextra")
install.packages("MASS")
library("MASS")
install.packages("ggplot2")
library("ggplot2")
install.packages("ggfortify")
library("ggfortify")
```

```r
Elements <- read.csv("allspec_shoot_xyz_concat.csv")
```

```r
#violinplot(Elements$B11_X~Elements$spec_as_int) #Boron 
Elements$spec_as_int <- as.factor(Elements$spec_as_int)
boron <- ggplot(Elements, aes(x = spec_as_int, fill = spec_as_int, alpha = .5)) + 
  geom_violin(aes(y = B11_X))+
  geom_violin(aes(y = B11_Y))+
  geom_violin(aes(y = B11_Z))+
  scale_color_viridis_d(aesthetics = "fill")+
  theme_bw()+
  xlab("Species")+
  ylab("Z-Score of Boron Concentration")+
  labs(title = expression(paste("Distribution of Boron Concentration among ", italic("Vitis "), "Species")))+ 
  guides(alpha = FALSE)+
  theme(plot.title = element_text(hjust = .55), legend.text = element_text(face = "italic"))+
  scale_fill_discrete(name = "Species", labels = c("V. acerifolia", "V. aestivalis", "V. cinera", "V. labrusca", "V. palmatta", "V. riparia","V. rupestris", "V. vulpina"))                
boron
ggsave(filename = "Boron.jpg", plot = boron, dpi = "retina")
```
