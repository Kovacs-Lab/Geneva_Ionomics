```r
install.packages("ggplot2")
library("ggplot2")
```

```r
Vitis <- read.csv("allspec_shoot_xyz_concat.csv")
Vitis$spec_as_int <- as.factor(Vitis$spec_as_int)
names = colnames(Vitis)[4:60]
```

```r
acerifolia <- subset(Vitis, spec_as_int == 1)
acerifoliaAvg <- c(4:60)
for (x in acerifoliaAvg){
    acerifoliaAvg[x-3] = mean(acerifolia[ ,x])
}
#barplot(acerifoliaAvg) #We are not using z scores to make it easier to see differences
df.acerifoliaAvg <- data.frame(acerifoliaAvg)
df.acerifoliaAvg <- data.frame(cbind(names,acerifoliaAvg))
acerifolia <- ggplot(df.acerifoliaAvg, aes(
        x = names,
        y = acerifoliaAvg,
        fill = factor(names)
)) + 
geom_col(width = 1, color = "white")
acerifolia <- acerifolia + coord_polar()
acerifolia <- acerifolia + labs(title = expression(paste(italic("V. acerifolia "), "leaf ionome")))
acerifolia <- acerifolia + theme(axis.title.x = element_blank(),
                    axis.title.y = element_blank(),
                    axis.text.y = element_blank(),
                    axis.ticks.y = element_blank(),
                    legend.position = 'none',
                    plot.title = element_text(hjust = .5),
                    axis.text.x=element_text(size=5))
acerifolia <- acerifolia + geom_hline(yintercept = 10,alpha = .3)
acerifolia <- acerifolia + geom_hline(yintercept = 20,alpha = .3)
acerifolia <- acerifolia + geom_hline(yintercept = 30,alpha = .3)
acerifolia <- acerifolia + geom_hline(yintercept = 40,alpha = .3)
acerifolia <- acerifolia + geom_hline(yintercept = 50,alpha = .3)
acerifolia <- acerifolia + geom_hline(yintercept = 60,alpha = .3)
acerifolia
ggsave(filename = "acerifolia.jpg", plot = acerifolia, dpi = "retina")
```
