```python
import pandas as pd
import seaborn as sns
import os
import matplotlib.pyplot as plt
sns.set(style="whitegrid",font_scale=2)
import matplotlib.collections as clt
import ptitprince as pt
```

```python
savefigs = False
figs_dir = '../figs'

if savefigs:
    # Make the figures folder if it doesn't yet exist
    if not os.path.isdir('../figs'):
        os.makedirs('../figs')
```

```python
def export_fig(axis,text, fname):
    if savefigs:
        axis.text()
        axis.savefig(fname, bbox_inches='tight')
```

```python
dfx = pd.read_csv ("gen_X_raw_ForRatios.csv", sep= ",")
dfy = pd.read_csv ("gen_Y_raw_ForRatios.csv", sep= ",")
dfz = pd.read_csv ("gen_Z_raw_ForRatios.csv", sep= ",")
```

```python
acerifolia = dfx[dfx["species_position"] == "acerifolia_x"].append(dfy[dfy["species_position"] == "acerifolia_y"]).append(dfz[dfz["species_position"] == "acerifolia_z"])
aestivalis = dfx[dfx["species_position"] == "aestivalis_x"].append(dfy[dfy["species_position"] == "aestivalis_y"]).append(dfz[dfz["species_position"] == "aestivals_z"])
cinerea = dfx[dfx["species_position"] == "cinerea_x"].append(dfy[dfy["species_position"] == "cinerea_y"]).append(dfz[dfz["species_position"] == "cinerea_z"])
labrusca = dfx[dfx["species_position"] == "labrusca_x"].append(dfy[dfy["species_position"] == "labrusca_y"]).append(dfz[dfz["species_position"] == "labrusca_z"])
palmatta = dfx[dfx["species_position"] == "palmatta_x"].append(dfy[dfy["species_position"] == "palmatta_y"]).append(dfz[dfz["species_position"] == "palmatta_z"])
riparia = dfx[dfx["species_position"] == "riparia_x"].append(dfy[dfy["species_position"] == "riparia_y"]).append(dfz[dfz["species_position"] == "riparia_z"])
rupestris = dfx[dfx["species_position"] == "rupestris_x"].append(dfy[dfy["species_position"] == "rupestris_y"]).append(dfz[dfz["species_position"] == "rupestris_z"])
vulpina = dfx[dfx["species_position"] == "vulpina_x"].append(dfy[dfy["species_position"] == "vulpina_y"]).append(dfz[dfz["species_position"] == "vulpina_z"])
```

```python
f, ax = plt.subplots(figsize=(7, 7))

sns.barplot(x = "species_position", y = "B11", data = acerifolia, capsize= .1)
plt.title("Figure P1\n Bar Plot")

if savefigs:
    plt.savefig('../figs/figureP01.png', bbox_inches='tight')
```

```python
f, ax = plt.subplots(figsize=(7, 5))

pt.RainCloud(x = "species_position", y = "B11", data = acerifolia, palette = "Set2", bw = .25,
                 width_viol = .6, ax = ax, orient = "h",move = .2, jitter = 1)

plt.title("Figure P7\n Using the pt.Raincloud function")
if savefigs:
    plt.savefig('../figs/tutorial_python/figureP07.png', bbox_inches='tight')
```
